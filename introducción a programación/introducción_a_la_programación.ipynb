
# ¿Qué es un algoritmo?
conjunto ordenado y finito de operaciones simples a través del cual podemos hallar la solución a un problema determinado y estipulado de ante mano.

La palabra algoritmo proviene del latín  *alborarismus*, y este a su vez es una abreviación del árabe *ḥisābu lḡubār*, que significa ‘cálculo mediante cifras arábigas’.

Los algoritmos  ejecutan instrucciones definidas, ordenadas y finitas. Así, dado un estado inicial y una entrada, y siguiendo los sucesivos pasos indicados, se llega al estado final y se obtiene una solución.
